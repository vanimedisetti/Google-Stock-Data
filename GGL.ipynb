## Step 7: Outlier Detection
Identify and handle outliers using visualizations and statistical methods.



```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Close'])
plt.title('Box Plot for Closing Prices')
plt.xlabel('Closing Price')
plt.show()
q1 = df['Close'].quantile(0.25)
q3 = df['Close'].quantile(0.75)
iqr = q3 - q1
outliers = df[(df['Close'] < (q1 - 1.5 * iqr)) | (df['Close'] > (q3 + 1.5 * iqr))]
print("Outliers in Closing Prices:")
print(outliers)
df_no_outliers = df[~df.index.isin(outliers.index)]
print("\nDataset after removing outliers:")
print(df_no_outliers.describe())
```

The box plot visualizes the distribution of Google's closing stock prices.

It reveals a right-skewed distribution with several outliers on the higher end.

The median closing price appears around 0, while the interquartile range (IQR) suggests a moderate spread.

The whiskers extend to the minimum and maximum non-outlier values, indicating potential volatility in the stock'sperformance.


1. **Box Plot for Outlier Detection**:
   - Creates a box plot to visualize the distribution of `Closing Prices` and identify potential outliers.
   - The box plot shows the range of prices, and points outside the "whiskers" could be outliers.

2. **Handling Outliers Using the IQR Method**:
   - Calculates the **Interquartile Range (IQR)** for `Closing Prices`, which is the difference between the 75th percentile (Q3) and the 25th percentile (Q1).
   - Defines outliers as values that are below `(Q1 - 1.5 * IQR)` or above `(Q3 + 1.5 * IQR)`.

3. **Display and Remove Outliers**:
   - Prints the rows of data that are considered outliers based on the IQR method.
   - Optionally, removes the outliers from the dataset and displays the updated dataset without outliers.

This program helps identify and handle extreme values (outliers) in the `Closing Prices` of the Google stock. Removing outliers can improve the accuracy of analysis and models by focusing on typical price trends.

## Step 8: Hypothesis Testing
Formulate hypotheses based on initial observations and test them using statistical methods.



```python
from scipy.stats import ttest_ind
import pandas as pd
df['Date'] = pd.to_datetime(df['Date'])
specific_date = df['Date'].median()
before_date = df[df['Date'] < specific_date]['Open']
after_date = df[df['Date'] >= specific_date]['Open']
t_stat, p_value = ttest_ind(before_date, after_date)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
alpha = 0.05 
```

 this code snippet demonstrates a common approach to hypothesis testing in statistical analysis, applied to Google stock data.
 

1. **Convert 'Date' to Datetime**:
   - The program ensures that the `Date` column is in the correct datetime format for easy comparison.

2. **Define a Specific Date**:
   - It selects a specific date (in this case, the median date of the dataset) to split the data into two groups: one for dates before and another for dates after the chosen date.

3. **Split Data into Two Groups**:
   - The program separates the `Open` prices into two groups: `before_date` (data before the specific date) and `after_date` (data after the specific date).

4. **Perform Independent t-test**:
   - A **t-test** is performed to determine if there is a significant difference between the opening prices before and after the specific date.

5. **Print Results**:
   - The t-statistic and p-value are printed to help assess if the difference in opening prices is statistically significant.

6. **Interpretation**:
   - The program compares the p-value to a significance level (alpha = 0.05) to conclude whether there is a significant change in the stock's opening price before and after the specific date.


The program helps determine if there's a statistically significant difference in the opening prices of Google stock before and after a certain date, which could be useful for identifying key events or trends.

**This program processes and visualizes the average opening prices of Google stock by year using the 2024 dataset:**

```python
import matplotlib.pyplot as plt
import pandas as pd
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
yearly_avg_open = df.groupby('Year')['Open'].mean()

plt.figure(figsize=(12, 7))

plt.bar(yearly_avg_open.index, yearly_avg_open.values, color='cornflowerblue', edgecolor='black')

plt.xlabel("Year", fontsize=12, labelpad=10)
plt.ylabel("Average Opening Price (USD)", fontsize=12, labelpad=10)
plt.title("Yearly Average Opening Prices of Google Stock", fontsize=14, pad=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for year, value in zip(yearly_avg_open.index, yearly_avg_open.values):
    plt.text(year, value + 1, f"{value:.2f}", ha='center', va='bottom', fontsize=10)

plt.xticks(yearly_avg_open.index, rotation=45, fontsize=10)

plt.tight_layout()  # Adjust layout to prevent overlapping text
plt.show()
```
